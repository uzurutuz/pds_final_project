```python
# Keeping only the relevant columns
col_to_keep=['PROPERTYZIP', 'SALEPRICE','PROPERTYSTATE','HEATINGCOOLINGDESC','BEDROOMS','TOTALROOMS',\
             'PROPERTYADDRESS','BASEMENT','PROPERTYCITY','DEEDBOOK','STYLE','CDUDESC','CONDITIONDESC',\
             'FULLBATHS','FIREPLACES','FINISHEDLIVINGAREA','HOMESTEADFLAG','LOTAREA','GRADE']

pr_val_cleaned = pr_val[col_to_keep]
```

```python
# Dropping the columns with Nan values
pr_val_clean = pr_val_cleaned.dropna(how='any')

# Mapping two features to make them numrical. There is an inherent order in the mapping
cdu_cond_mapping = {'EXCELLENT':8,'VERY GOOD':7,'GOOD':6,'FAIR':5,'AVERAGE':4,'POOR':3,'VERY POOR':2,'UNSOUND':1}
pr_val_clean['CDUDESC'] = pr_val_clean['CDUDESC'].map(cdu_cond_mapping)
pr_val_clean['CONDITIONDESC'] = pr_val_clean['CONDITIONDESC'].map(cdu_cond_mapping)
```

```python
unique_properties = pr_val_clean['HEATINGCOOLINGDESC'].unique()

# Mapping the feature as there is an inherent order.
heat_cool_mapping = {'Central Heat with AC':15,'Heat Pump with AC':14,'Electric Heat with AC':13,'Unit Furnace with AC':12,\
                 'Floor Furnace with AC':11,'Wall Furnace with AC':10,'Central Heat':9,'Heat Pump':8,'Electric':7,\
                 'Unit Heat':6,'Floor Furnace':5,'Wall Furnace':4,'Other':3,'No heat but with AC':2,'None':1}

pr_val_clean['HEATINGCOOLINGDESC'] = pr_val_clean['HEATINGCOOLINGDESC'].map(heat_cool_mapping)
```

```python
pr_val_tmp = pr_val_clean.copy()

# Considering only those rows whose state is Pennsylvania
if pr_val_clean['PROPERTYSTATE'].unique()=='PA':
    pr_val_clean = pr_val_tmp.drop(['PROPERTYSTATE'],axis=1)
else:
    pr_val_clean = pr_val_tmp[pr_val_tmp['PROPERTYSTATE']=='PA']
    pr_val_clean = pr_val_tmp.drop(['PROPERTYSTATE'],axis=1)
    
```

```python
# One hot encoding of categorical variables
temp_df = pd.get_dummies(pr_val_clean['PROPERTYCITY']).astype(np.int64)
new_pr = pd.concat([pr_val_clean,temp_df],axis=1)
pr_df = new_pr.copy()
```

```python
col_to_drop=['PROPERTYADDRESS','PROPERTYCITY','DEEDBOOK','STYLE']

if len(new_pr['HOMESTEADFLAG'].unique())==1:
    col_to_drop.append('HOMESTEADFLAG')
else:
    print('There are multiple unique values please check data')
    
try:
    pr_df = new_pr.drop(col_to_drop,axis=1)
except:
    print('An error has occured')
```

## Social Characteristics Data


```python
def get_median_categorical(size, df):
    
    df_cumulative = df
    
    median_size = size / 2

    for i in range(1, len(df_cumulative.columns)):
        df_cumulative.iloc[:, i] = df_cumulative.iloc[:, i - 1] + df_cumulative.iloc[:, i]

    cols = df_cumulative.columns
    df_cumulative['median'] = ''
    
    for i, row in df.iterrows():
        vals = np.array(row[:-1].values)
        indx = np.where(vals >= median_size[i])[0][0]
        df_cumulative.loc[i, 'median'] = cols[indx]
        
    return df_cumulative['median']
    pass

df_social = pd.read_csv("Data/social_characteristics_PIT/social_characteristics_PIT.csv")
df_social.columns =['zipcode', 'avg_household_size', 'pop_over_25', 'less_9th_grade', '9th_to_12th_grade', 'high_school_grad', 'some_college_no_degree', 'associate_degree', 'bachelor_degree', 'graduate_or_professional_degree']
#df_social = df_social.set_index('zipcode')
df_social

df_social['median_education'] = get_median_categorical(df_social['pop_over_25'], df_social[['less_9th_grade', '9th_to_12th_grade', 'high_school_grad', 'some_college_no_degree', 'associate_degree', 'bachelor_degree', 'graduate_or_professional_degree']])
df_social.head()
```

# Baseline Model
## Linear Regression vs Random Forest Models
We have applied both a [linear regression model](http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html) and [random forest model](http://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html) using the scikit learn package.


```python
from sklearn.preprocessing import StandardScaler
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

def get_params(data):
    # the X matrix and Y matrix

    y=data.iloc[:,1].values
    X=data.iloc[:,2:].values

    # Standardizing the data
    sc_x=StandardScaler()
    X_std=sc_x.fit_transform(X)
    
    # Splitting data set into training and test set
    
    X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.3,random_state=1)
    
    return X_train,X_test,Y_train,Y_test
    
def linear_reg(X_train,X_test,Y_train,Y_test):
    
    # Running the linear regression algorithm.
    slr=LinearRegression()
    slr.fit(X_train,Y_train)

    R_squared_error=slr.score(X_test,Y_test)

    print('The R^2 error is found to be {0:3f} for Linear Regression'.format(R_squared_error))
    
    return R_squared_error
    
def random_forest(X_train,X_test,Y_train,Y_test):
    regr = RandomForestRegressor(max_depth=7, random_state=0)
    regr.fit(X_train,Y_train)

    R_squared_error = regr.score(X_test,Y_test)

    print('The R^2 error is found to be {0:3f} for Random Forest Regression'.format(R_squared_error))
    
    return R_squared_error
```

```python
X_train,X_test,Y_train,Y_test = get_params(pr_df)
Rs_lr = linear_reg(X_train,X_test,Y_train,Y_test)
Rs_rfr = random_forest(X_train,X_test,Y_train,Y_test)
```

```python
# Remove Outliners that do not make sense
pr_df_re = pr_df.drop(pr_df[pr_df['SALEPRICE'] < 20000].index)

X_train,X_test,Y_train,Y_test = get_params(pr_df_re)
Rs_lr_ = linear_reg(X_train,X_test,Y_train,Y_test)
Rs_rfr_ = random_forest(X_train,X_test,Y_train,Y_test)
```

Functions used to generate points inside each polygon shape representing numerical features when plotting.

```python

import geopandas as gpd
from shapely.geometry import shape  
import pandas as pd
import requests
from shapely.geometry import Point
from numpy.random import RandomState, uniform
import numpy as np

# START REFERENCED CODE
# Code aquired from https://github.com/agaidus/census_data_extraction/blob/master/census_mapper.py
# Author: Andrew Gaidus
def gen_random_points_poly(poly, num_points, seed = None):
    """
    Returns a list of N randomly generated points within a polygon. 
    """
    min_x, min_y, max_x, max_y = poly.bounds
    points = []
    i=0
    while len(points) < num_points:
        s=RandomState(seed+i) if seed else RandomState(seed)
        random_point = Point([s.uniform(min_x, max_x), s.uniform(min_y, max_y)])
        if random_point.within(poly):
            points.append(random_point)
        i+=1
    return points


def gen_points_in_gdf_polys(geometry, values, points_per_value = None, seed = None):
    """
    Take a GeoSeries of Polygons along with a Series of values and returns randomly generated points within
    these polygons. Optionally takes a "points_per_value" integer which indicates the number of points that 
    should be generated for each 1 value.
    """
    if points_per_value:
        new_values = (values/points_per_value).astype(int)
    else:
        new_values = values
    new_values = new_values[new_values>0]
    g = gpd.GeoDataFrame(data = {'vals':new_values}, geometry = geometry)
    
    a = g.apply(lambda row: tuple(gen_random_points_poly(row['geometry'], row['vals'], seed)),1)
    b = gpd.GeoSeries(a.apply(pd.Series).stack(), crs = geometry.crs)
    b.name='geometry'
    return b

# END REFERENCED CODE

def dot_density_map(data, col_names, num_people=150):

    list_of_point_categories = []
    for col in col_names:
        ps = gpd.GeoDataFrame(gen_points_in_gdf_polys(data['geometry'], data[col], num_people, 10))
        
        ps['field'] = col
        list_of_point_categories.append(ps)

    return gpd.GeoDataFrame(pd.concat(list_of_point_categories))
```

```python
df_saleprice = pd.DataFrame(pr_val_cleaned['PROPERTYZIP'].unique())
df_saleprice = df_saleprice.rename(columns={0:'PROPERTYZIP'})
df_saleprice = df_saleprice.dropna()


list_label = ['less_150K', '150K_to_300K', '300K_to_450K','450K_to_600K','600K_or_more']

for i in range(len(list_range) + 1):
    df_sales = pr_val_cleaned[['PROPERTYZIP', 'SALEPRICE']]
    if i == 0:
        df_temp = df_sales[df_sales['SALEPRICE'] < list_range[i]].groupby('PROPERTYZIP').count().reset_index()
        df_saleprice = pd.merge(df_saleprice, df_temp, on='PROPERTYZIP', how='left')

    elif i == len(list_range): 
        df_temp = df_sales[df_sales['SALEPRICE'] >= list_range[-1]].groupby('PROPERTYZIP').count().reset_index()
        df_saleprice = pd.merge(df_saleprice, df_temp, on='PROPERTYZIP', how='left')

    else:
        df_temp = df_sales[(df_sales['SALEPRICE'] >= list_range[i-1]) & (df_sales['SALEPRICE'] < list_range[i])].groupby('PROPERTYZIP').count().reset_index()
        df_saleprice = pd.merge(df_saleprice, df_temp, on='PROPERTYZIP', how='left')
        


df_saleprice = df_saleprice.fillna(0)
def df_column_uniquify(df):
    '''To solve the problem of columns having similar names, the following function is used.'''
    # https://stackoverflow.com/questions/24685012/pandas-dataframe-renaming-multiple-identically-named-columns
    df_columns = df.columns
    new_columns = []
    for item in df_columns:
        counter = 0
        newitem = item
        while newitem in new_columns:
            counter += 1
            newitem = "{}_{}".format(item, counter)
        new_columns.append(newitem)
    df.columns = new_columns
    return df
df_saleprice = df_column_uniquify(df_saleprice)
for i in range(len(list_range) + 1):
    if i == 0:
        df_saleprice = df_saleprice.rename(columns={ df_saleprice.columns[i+1]: list_label[i] })
    elif i == len(list_range):
        df_saleprice = df_saleprice.rename(columns={ df_saleprice.columns[i+1]: list_label[i] })

    else:
        df_saleprice = df_saleprice.rename(columns={ df_saleprice.columns[i+1]: list_label[i] })

df_saleprice = df_saleprice.rename(columns={ df_saleprice.columns[0]: "zipcode" })

df_saleprice.head()
```

# Combine Assessment Data with Socio-Economic and Demographic Data

## Generate Socio-Economic and Demographic Features
We will start by merging the new features into a single DataFrame. The features we will be including are:
- Median Household Income
- Majority Race (White or Balck)
- Median Education
<br>These features will be merged using the zipcode as a key.

```python
df_economic = df_economic[df_economic.median_household_income != '-']

df_features = pd.merge(df_economic[['zipcode', 'median_household_income']],\
                           df_demographic[['zipcode', 'majority_race']], how='inner', on='zipcode', copy=True)
df_features = pd.merge(df_features,\
                           df_social[['zipcode', 'median_education']], how='inner', on='zipcode', copy=True)

df_features['median_household_income'] = df_features['median_household_income'].astype('float64')
df_features.head()
```

The two of the resulting features are categorical, therefore, we will use one-hot-encoding to convert those categorical columns into features we can use in our model.

```python
dummies_race = pd.get_dummies(df_features['majority_race']).astype('float64')
dummies_education = pd.get_dummies(df_features['median_education']).astype('float64')

df_new_features = df_features[['zipcode', 'median_household_income']]
df_new_features = pd.concat([df_new_features, dummies_race, dummies_education], axis=1)

df_new_features.head()
```

## Merge the Initial Features and Dependent Variable with the New Features
The new features will be merged with the original assessment features using the zip code, so that addresses in the same zip code will share the new features.

```python
pr_df['PROPERTYZIP'] = pr_df['PROPERTYZIP'].astype(np.int64)

# Merge by zipcode
df_final = pd.merge(pr_df, df_new_features, how='inner', left_on='PROPERTYZIP', right_on='zipcode', copy=True)

df_final = df_final.drop(['zipcode'], axis=1)
df_final.head()
```

## Applied Models
Now we will apply both linear regression and random forest regression to the new features.

```python
X_train,X_test,Y_train,Y_test = get_params(df_final)
Rs_lr_new = linear_reg(X_train,X_test,Y_train,Y_test)
Rs_rfr_new = random_forest(X_train,X_test,Y_train,Y_test)
```

As it can be seen, the R-Squares improves a little bit (just 0.03 for LR and 0.04 for RFR) and in this case RFR, presents better results. The influence of race is not significant to property value. This is probably due to the low diversity of race in Allegheny County, as most of the zip codes are white adding little value to the model. Despite this fact, we will leave the feature inside the model as it might be interesting to see if this is different for other counties or cities with more diversity.

After this improvement, we will also try to applied the previous trick of eliminating the rows with sales prices that are below what we will expect (<$20,000).

```python
# Remove Outliners that do not make sense
df_final_re = df_final.drop(df_final[df_final['SALEPRICE'] < 20000].index)

X_train,X_test,Y_train,Y_test = get_params(df_final_re)
Rs_lr_new_ = linear_reg(X_train,X_test,Y_train,Y_test)
Rs_rfr_new_ = random_forest(X_train,X_test,Y_train,Y_test)
```

Unlike with the initial baseline model, this trick improves considerably the accuracy of fit, achieving R-Squares above 0.5 for both LR and RFR.
